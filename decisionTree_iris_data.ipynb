```python
from sklearn import tree, metrics,datasets
import numpy as np
import matplotlib.pyplot as plt
from sklearn import model_selection

data = datasets.load_iris()
print(data.feature_names)
print(data.target)
```

```python
# divide the data into train and test 
X = data.data
y = data.target

Xtrain,Xtest,ytrain,ytest = model_selection.train_test_split(X,y,test_size=0.7,random_state=2)


```

```python
# create the tree model 
treeModel = tree.DecisionTreeClassifier(criterion='gini') # 'gini' , 'entropy'
# train the DT 
treeModel = treeModel.fit(Xtrain,ytrain)
# chk the training accuracy and testing accurary 

YtrainPred = treeModel.predict(Xtrain)
YtestPred = treeModel.predict(Xtest)

# training accuracy 
trainAcc = metrics.accuracy_score(YtrainPred,ytrain)
# testing accuracy 
testAcc = metrics.accuracy_score(YtestPred,ytest)

print("DT Training Accuracy: ",trainAcc)
print("DT Testing Accuracy: ",testAcc)

```

```python
# visualise the tree 
plt.figure(figsize=(15,12))
tree.plot_tree(treeModel,rounded=True,filled=True)
```

```python
conf = metrics.confusion_matrix(YtestPred,ytest)
```

```python
report = metrics.classification_report(YtestPred,ytest)
```

```python
import pandas as pd 

path = './diabetes.csv'

dat = pd.read_csv(path)

X = dat.drop('Outcome',axis=1).values
y = dat['Outcome'].values
print(X.shape,y.shape)

```

```python
Xtrain,Xtest,ytrain,ytest = model_selection.train_test_split(X,y,test_size=0.7,random_state=2)

# create the tree model 
treeModel = tree.DecisionTreeClassifier(criterion='gini') # 'gini' , 'entropy'
# train the DT 
treeModel = treeModel.fit(Xtrain,ytrain)
# chk the training accuracy and testing accurary 

YtrainPred = treeModel.predict(Xtrain)
YtestPred = treeModel.predict(Xtest)

# training accuracy 
trainAcc = metrics.accuracy_score(YtrainPred,ytrain)
# testing accuracy 
testAcc = metrics.accuracy_score(YtestPred,ytest)

print("DT Training Accuracy: ",trainAcc)
print("DT Testing Accuracy: ",testAcc)


```

```python
from sklearn import ensemble
Xtrain,Xtest,ytrain,ytest = model_selection.train_test_split(X,y,test_size=0.5,random_state=2)

# create the tree model 
treeModel = tree.DecisionTreeClassifier(criterion='entropy') # 'gini' , 'entropy'

RFModel = ensemble.RandomForestClassifier(n_estimators=100,criterion='entropy') # 'gini' , 'entropy'

# train the DT 
treeModel = treeModel.fit(Xtrain,ytrain)
RFModel = RFModel.fit(Xtrain,ytrain)
# chk the training accuracy and testing accurary 

YtrainPred = treeModel.predict(Xtrain)
YtrainPredRF = RFModel.predict(Xtrain)

YtestPred = treeModel.predict(Xtest)
YtestPredRF = RFModel.predict(Xtest)

# training accuracy 
trainAcc = metrics.accuracy_score(YtrainPred,ytrain)
trainAccRF = metrics.accuracy_score(YtrainPredRF,ytrain)
# testing accuracy 
testAcc = metrics.accuracy_score(YtestPred,ytest)
testAccRF = metrics.accuracy_score(YtestPredRF,ytest)


print("DT Training Accuracy: ",trainAcc)
print("DT Testing Accuracy: ",testAcc)

print("RF Training Accuracy: ",trainAccRF)
print("RF Testing Accuracy: ",testAccRF)




```
